```python
# GC Content em percentagem
# https://biopython.org/docs/1.75/api/Bio.SeqUtils.html
from Bio.SeqUtils import GC

gc_cont=[]
for x in full_ds.capsular_bacterial_DNA:
    gc_cont.append(GC(x))
```

```python
# AT Content
# https://stackoverflow.com/questions/16908475/beginner-python-script-for-calculating-gc-content-in-dna-sequence

at_cont=[]
for x in full_ds.capsular_bacterial_DNA:
    total = len(x)
    a = x.count("A")
    t = x.count("T")
    at_total = a+t
    at_content = (at_total/float(total))*100
    at_cont.append(at_content)

table_at=pd.DataFrame({'AT_Content_DNA':at_cont})

full_ds=full_ds.join(table_at)
full_ds
```

```python
'''
Kmers for genome comparison
'''

def kmer_gen(seq, k):
    """
    Generates all kmers of length k from a sequence
    """
    kmers = []
    for i in range(len(seq) - k + 1):
        kmers.append(seq[i:i + k])
    return kmers

# obtido através do github copilot
```

```python
kmer_gen(full_ds.capsular_bacterial_DNA[0],4)
```

```python
# kmers

kmers=[]
for k in full_ds.capsular_bacterial_DNA:
    kmers.append(kmer_gen(k,4))

kmers
```

```python
table_kmers=pd.DataFrame({'Kmers':kmers})
table_kmers
```
